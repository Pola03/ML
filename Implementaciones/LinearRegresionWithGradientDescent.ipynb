```python
import pandas as pd
import numpy as np

# Descargar el conjunto de datos desde su fuente original
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# Preparar los datos
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

print("Datos (features):", data.shape)
print("Objetivo (target):", target.shape)
```

Para utilizar el descenso del gradiente para ajustar un modelo de regresión lineal, necesitamos minimizar la función de costo asociada al modelo. La función de costo típica para la regresión lineal es el error cuadrático medio (MSE):

$
J(\mathbf{B}) = \frac{1}{2m} \sum_{i=1}^m \left( y_i - \hat{y}_i \right)^2 = \frac{1}{2m} \sum_{i=1}^m \left( y_i - \mathbf{B}^T \mathbf{x}_i \right)^2
$

Donde:
- $\mathbf{B} = [b_0, b_1]$ son los parámetros del modelo (intercepto y pendiente).
- $\mathbf{x}_i = [1, x_i]$ son los datos de entrada, con un término constante $1$ para el intercepto.
- $y_i$ son las etiquetas verdaderas.
- $m$ es el número de datos.

El descenso del gradiente actualizará los parámetros $\mathbf{B}$ iterativamente en la dirección opuesta al gradiente de $J(\mathbf{B})$, con una tasa de aprendizaje $lr$.


2. **Gradiente**:
   - El gradiente se calcula como:
     $
     \nabla J(\mathbf{B}) = \frac{1}{m} \mathbf{X}^T (\mathbf{X} \mathbf{B} - \mathbf{Y})
     $
   - Este gradiente se utiliza para actualizar los parámetros en la dirección opuesta.

3. **Costo**:
   - Se calcula el costo $J(\mathbf{B})$ en cada iteración para verificar la convergencia.

```python
X = np.array(data[:,5]) 
Y = np.array(target)
# print(X.shape)
# print(X)
# print(Y.shape)
# print(Y)
plt.scatter(X,Y, alpha=0.5)

#Anadir columna de unos para el termino independiente
X = np.array([np.ones(506),X]).T
#506 arrays con dos elementos
# print(X)

# print(X.shape)

B = np.linalg.inv(X.T @ X) @ X.T @ Y

plt.plot([4,9],[B[0] + B[1]* 4, B[0] + B[1] *9], c = "red")

plt.show()
```
